```python
#Plots are displayed inline
%matplotlib inline
```

```python
#Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#Set the x axis months.
plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
plt.subplots(figsize=(20,10))
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#Set the x axis months.
plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Object-oriented Approach
#Create the plot.
ax = plt.axes()
ax.plot(x_axis, y_axis, color="green", marker= "d", linewidth=2, label="Boston")
#Add a legend.
ax.legend()
#Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#Add a y-axis limit
ax.set_ylim(0,45)
#Add grid lines
ax.grid()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label = "Boston")
plt.legend()
#Add a title and axes labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
#Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
#Invert the y-axis data.
ax.invert_yaxis()
```

```python
#Create a Scatter Plot
plt.plot(x_axis, y_axis, 'o')
```

```python
#Create a Scatter Plot
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
#Add a legend.
plt.legend()
#Add a title and axes labels.
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
#Add a limit for the x-axis data.
plt.xlim(0,50)
#Invert the y-axis so that January is at the top. 
plt.gca().invert_yaxis()
```

```python
#Create a Bubble Chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Adjust the marker size
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

#Adjusted scatter plot
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Use list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Scatter Plot Object-Oriented Method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#Bubble Plot Object-Oriented Method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Bubble Plot Object-Oriented Method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis],color="skyblue",edgecolors="black", linewidth=2,  label="Boston", alpha = 0.8)
#Add a legend.
ax.legend()
#Add a title.
ax.set_title("PyBer Fare by Month")
#Add a limit for the x-axis.
ax.set_xlim(0,50)
#Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

```python
#Create a pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#Explode or pop out one wedge
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#Create a pie chart using the object-oriented method.
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis,autopct='%.1f%%',explode=explode_values,shadow=True,colors=colors,startangle=90,counterclock=False)
```
